##### Grafica 1: Diagrama de barras (axe[0,0])

```python
#Primero filtramos el dataframe por las ciudades que queremos mostrar en la gráfica. Ojo, los nombres no tienen separación:
ciudades_mas_pobladas = df_california["city"].isin(["LosAngeles", "SanDiego", "Sacramento", "Anaheim", "SantaAna"])
df_filtro_ciudades = df_california[ciudades_mas_pobladas]
df_filtro_ciudades
```

```python
#Creamos la gráfica para mostrar la frecuencia absoluta, mostrando cada ciudad como categoría:

categorias = df_filtro_ciudades["city"].value_counts().index.to_list()
frecuencia_absoluta = df_filtro_ciudades["population_total"].value_counts().to_list()

fig1 = plt.figure(figsize=(7,6)) #Se crea el contenedor figura
ax1 = plt.axes() #Se crea el axe donde irá la gráfica

ax1.bar(categorias, frecuencia_absoluta, color= ["blue","grey", "red", "olive", "brown"], alpha = 0.5)
ax1.set_title("Ciudades más pobladas de california")
ax1.set_xlabel("Ciudades")
ax1.set_ylabel("Habitantes (miles)")
ax1.set_xticklabels(["Los Angeles", "San Diego", "Sacramento", "Anaheim", "Santa Ana"]);

#Hay que ajustar el eje "y" y añadir texto con los límites de cada barra --> plt.text()
```

##### Grafica 2: Histograma (axe[0,1])

```python
data = df_california["area_total_km2"]

fig2 = plt.figure()
ax2 = plt.axes()
ax2.hist(data, bins=50, color="steelblue", alpha=0.5, label="Hist Area")

font_dict = {"size": 18,
             "fontfamily": "Bradley Hand itc"} #Diccionario para el estilo de tipografía de las etiquetas

ax2.set_xlabel("Superficie $(km^2)$", fontdict= font_dict)
ax2.set_ylabel("Numero de ciudades", fontdict= font_dict)
ax2.legend();
```

##### Grafica 3: Scatterplot (axe[1,0])

```python
fig3 = plt.figure(figsize= (7,5))
ax3 = plt.axes()

pob, area = df_california["population_total"], df_california["area_water_km2"]

ax3.scatter(area, pob, c="green", alpha= 0.5)

#Falta ajustar los valores del eje y .set_yticks()?

ax3.set_ylabel("Población(miles de habitantes)")
ax3.set_xlabel("Area de agua en la ciudad $(km^2)$")
ax3.set_title("Población por kilómetro cuadrado de agua")

props_flecha = {
    "arrowstyle": "<->",
    "connectionstyle": "angle3"
} #Diccionario para el estilo de flecha

ax3.annotate("Outliers", xy= (80.0, 3000.0), arrowprops= props_flecha)
```

##### Grafica 4: Diagrama de cajas (axe[1,1])

```python
#Vemos el valor del IQR para sacar el valor de "whis" en el boxplot:
df_california["elevation_m"].describe()
iqr = 173 - 22 

altitud = df_california["elevation_m"]
fig4 = plt.figure(figsize=(8,6))
ax4= plt.axes()
ax4.boxplot(altitud, whis=153, vert=False)
ax4.set_xlim(0,2500)
ax4.set_xlabel("Altitud (m)")
ax4.set_title("Diagrama de caja de la Altitud de las ciudades de California");
```
